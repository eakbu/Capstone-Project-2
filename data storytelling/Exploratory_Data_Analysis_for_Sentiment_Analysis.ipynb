# Amazon Health and Personal-Care Products Review-Sentiment Analysis

###### Problem statement
Health and personal care products are increasingly important and even vital for people’s health after the COVID-19 outbreak. What customers think about how effective these products in making us clean and healthy has not been investigated, however. In this project, my client is Amazon, and I will provide a software tool to Amazon in a way I will identify positive and negative words included in customers’ reviews for health and personal care products. For this purpose, I will use the customer reviews on these products in the period of 2004-2014. Based on these reviews in the said time period, I will develop a sentiment analysis model through natural language processing to determine customers’ sentiment toward these products. By developing such a model, I will provide my client, Amazon, better understanding of customers’ opinion and thoughts about its health and personal care products. **Thus, the central purpose of this project is to collect and analyze customer feedback about Amazon’s health and personal care products, and provide Amazon the detailed analyses of customer impressions on these products so that Amazon can benefit from these analyses in order to make improvements on its products and maximize e-commerce sales.**
###### Description of the Data
The data includes features in relation to customer reviews and ratings on health and personal care products. The data has 54787 rows and 9 features. In each row, I have a customer review and the following variables:



* reviewerID : ID of the reviewer, type: object
* asin : ID of the product , type: object reviewerName : the reviewer’s 
* name – type: object helpful : s, how helpful the review is – type: 
* object reviewText : text of the review – type:
* object overall : Rating – type: float64
* summary : summary of the review – type: object
* unixReviewTime : time of the review (unix time) – type: int64
* reviewTime : time of the 
* review (raw) – type: object

The data is located in Standford Analysis Project webpage. The original data is in a JSON format. To analyze the data, I will have to change the data format. In order to do that, I will import a JSON file and decode JSON file by using query in order to convert JSON file to csv file.

##### How the number of rating classes  change for each year ?

```python
# Select the data frame

df_new=df[['year','rating_class','review_text']]

# One hot encoding in the dataframe 
df_dummy=pd.get_dummies(df_new,columns=['rating_class'])

# Groupby year column and aggregate number of rating class and total reviews for each year 
df_group=df_dummy.groupby('year').agg({'rating_class_bad':['sum','mean'],'rating_class_good':['sum','mean'],
                             'rating_class_neutral':['sum','mean'],'review_text':['size']})


# Rename columns name
df_group.columns=['_'.join(cols).strip() for cols in df_group.columns.values]
df_group=df_group.reset_index()

df_group['rating_class_bad_mean']=df_group['rating_class_bad_mean']*100
df_group['rating_class_good_mean']=df_group['rating_class_good_mean']*100
df_group['rating_class_neutral_mean']=df_group['rating_class_neutral_mean']*100



df_group.columns=['Year','Bad Ratings Total','% of Bad Rating to Total Review in This Year ',
                  'Good Ratings Total','% of Good Rating to Total Review in This Year ',
                  'Neutral Ratings Total','% of Neutral Rating to Total Review in This Year ','Total Reviews']

# Print number of rating class and total reviews for each year
print(df_group[['Year','Bad Ratings Total','Good Ratings Total','Neutral Ratings Total']])



```

<font color='green'>
    The table above demonstrates the distribution of bad, good and neutral ratings over time. It suggests that the number of ratings considerably increases regardless of the class of ratings. In 2008, we see a very dramatic increase in the number of each class rating. For example, the number of bad ratings increases from 28 to 125; the number of good ratings increases from 351 to 1057; and the number of neutral ratings increases from 50 to 233. We also see a similar jump in 2013, too. For example, the number of bad ratings more than doubled in 2013 compared to the bad ratings in 2012. The amount of increase is even higher in the number of good ratings as it rises from 4543 to 15883. 


Although there is a stable increase on the number of ratings on each class from 2004 to 2013, it reports that in 2014, the number of ratings significantly reduced on every class. For instance, the number of neutral and bad ratings reduced by more than 50 percent.</font>

```python
# Make a new dataframe with clean text and rating classes
df_new=df_dummy2[['clean_text','rating_class_bad']]
df_new.head()
```

```python
# Initialize the countervectorizer
countVec=CountVectorizer(binary=True)
```

```python
# Fit the 'clean_text' to countvectorizer
countVec.fit(df_new['clean_text'])
```

```python
# Transform the matrix
transformed_matrix= countVec.transform(df_new['clean_text'])
```

```python
# Convert matrix to array
transformed_matrix.toarray()
```

```python
# Extract the feature names
names= countVec.get_feature_names()

# Export it out one sample
names[250]
```

```python
# Adding a 'rating' column 

df_transf=pd.DataFrame(
    transformed_matrix.toarray(),
    columns=names)


df_transf['rating_bad']=df_new['rating_class_bad']


```

```python
# Create lists for forming a dataframe summary
feature_names = []
avg_bad_ratings = [] 
bad_rating_counts = []

for name in names:
    if name != 'rating_bad':    
        avg_bad_rating   = df_transf[df_transf[name]== 1]['rating_bad'].mean()
        bad_rating_count = df_transf[df_transf[name]== 1]['rating_bad'].count()
        feature_names.append(name)
        avg_bad_ratings.append(avg_bad_rating)
        bad_rating_counts.append(bad_rating_count)  
  
    
    else:
        pass    
    
```

```python
# Create a new dataframe from words, average ratings, and rating counts
df_summary_bad = pd.DataFrame(
    {'feature_name':feature_names, 
     'avg_bad_rating': avg_bad_ratings, 
     'bad_rating_count':bad_rating_counts,})
     
```

```python
# Let's see the new dataframe
df_summary_bad.sample(10)
```

```python
# Words that are commonly used in the reviews which have bad ratings
df_bad = df_summary_bad.query("bad_rating_count > 100").sort_values(by='avg_bad_rating', ascending=False)[:100]
```

```python
df_bad.head(50)
```

```python
wc_bad = dict(zip(df_bad['feature_name'].tolist(), df_bad['avg_bad_rating'].tolist()))
```

```python
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                min_font_size = 10).generate_from_frequencies(wc_bad)
 
# plot the WordCloud image                       
plt.figure(figsize = (10, 10), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```

<font color='green'>
The most common 50 words in bad reviews are shown in the table above. Each of these words indicate the customers' impression on the products. For example, "ineffective" word suggests that the customer thinks that the product is not effective in cleaning or protecting the customer's health. According to the table, it is possible to say that the word "disappointment" has been most used by the customers in their bad reviews, suggesting that those customers found that the products they purchased fail to clean or protect the customers' health. </font>
